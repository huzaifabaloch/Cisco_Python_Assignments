# Marksheet

```python
import pandas as pd

names = []
roll_numbers = []
english_marks = []
math_marks = []
programming_fund_marks = [] 
automata_marks = []
data_structure_marks = []
total_markss = []
percentages = []
grades = []
results = {}

def check_if_roll_no_exist(roll_no):
    """Check if roll number entered is not used by some other student."""
    
    if roll_no in roll_numbers:
        return True
    return False

def calulate_grade_and_percentage():
    """Calulate grade and percentage."""
    
    total_marks = english_marks[-1] + math_marks[-1] + programming_fund_marks[-1] + automata_marks[-1] + data_structure_marks[-1]
    percentage = (total_marks / 500) * 100
    
    if percentage >= 80 and percentage <= 100:
        grade = 'A+'
    elif percentage >= 70 and percentage < 80:
        grade = 'A'
    elif percentage >= 60 and percentage < 70:
        grade = 'B'
    elif percentage >= 50 and percentage < 60:
        grade = 'C'
    elif percentage >= 40 and percentage < 50:
        grade = 'D'
    elif percentage >= 30 and percentage < 40:
        grade = 'E'
    else:
        grade = 'F'
        
    return total_marks, percentage, grade
    
def add_student():
    """Adding a student record."""
    
    name = input('Enter Student Name: ')
    if name.isdigit():
        print('input name not number!')
        return
    names.append(name)
    
    while True:
        roll_no = int(input('Enter a unique roll number: '))
        flag = check_if_roll_no_exist(roll_no)
        if flag == True:
            print('Sorry Student already exist on that roll number.\nTry a different one.')
        else:
            roll_numbers.append(roll_no)
            break
         
    english = int(input('Enter english marks: '))
    english_marks.append(english)
    maths = int(input('Enter mathematics marks: '))
    math_marks.append(maths)
    programming_fund = int(input('Enter programming fundamentals marks: '))
    programming_fund_marks.append(programming_fund)
    automata = int(input('Enter automata marks: '))
    automata_marks.append(automata)
    data_structures = int(input('Enter data structures marks: '))
    data_structure_marks.append(data_structures)
    total_marks, percentage, grade = calulate_grade_and_percentage()
    total_markss.append(total_marks)
    percentages.append(percentage)
    grades.append(grade)
    results.update({'Roll Number': roll_numbers, 'Name': names, 'English': english_marks, 'Math': math_marks, \
                    'Programming Fundamentals': programming_fund_marks, 'Automata': automata_marks, \
                    'Data Structures': data_structure_marks, 'Total Marks': total_markss, 'Percentage': percentages,\
                    'Grade': grades})
    
def main():
    
    while True:
        print('\n')
        print('1. Add Student Record')
        print('2. See Student Result')
        print('3. Delete Student Record')
        print('4. Save Results')
        print('5. Exit')
        option = int(input())
    
        if option == 1:
            add_student()
            
        elif option == 2:
            if results:
                df = pd.DataFrame(results)
                print(df)
                print('\n\n')
            else:
                print('No Record Available!')
            
        elif option == 3:
            roll_no = int(input('Enter roll number: '))
            flag = check_if_roll_no_exist(roll_no)
            if flag:        
                index_no = roll_numbers.index(roll_no)
                for each in results.values():
                    del each[index_no]
                print('Student Removed..\n')
            else:
                print('No such roll number found!')
            
        elif option == 4:
            if results:
                df.to_csv('results.csv')
                print('File Saved as results.csv')
            else:
                print('No Record Available!')
                    
        else:
            break 
```

```python
main()
```
